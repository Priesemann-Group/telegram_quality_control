```python
import pandas as pd
from pathlib import Path
from datetime import datetime
```

```python
folder = Path("./data/pro_kremlin_dataset")

min_timestamp = datetime(2027, 1, 1)
max_timestamp = datetime(1900, 1, 1)

for file in folder.glob("*.csv"):
    print(file)
    df = pd.read_csv(file)
    df["post_datetime"] = pd.to_datetime(df["post_datetime"])
    df["post_datetime"] = df["post_datetime"].dt.tz_localize(None)
    min_timestamp = min(min_timestamp, df["post_datetime"].min())
    max_timestamp = max(max_timestamp, df["post_datetime"].max())
```

```python
folder = Path("./data/anti_kremlin_dataset")

for file in folder.glob("*.csv"):
    print(file)
    df = pd.read_csv(file)
    df["post_datetime"] = pd.to_datetime(df["post_datetime"])
    df["post_datetime"] = df["post_datetime"].dt.tz_localize(None)
    min_timestamp = min(min_timestamp, df["post_datetime"].min())
    max_timestamp = max(max_timestamp, df["post_datetime"].max())
```

```python
print(f"Min timestamp: {min_timestamp}, max timestamp: {max_timestamp}")
```
